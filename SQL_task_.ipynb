```python
import sqlite3
import pandas as pd

connection = sqlite3.connect("workforce_analytics.db")
cursor = connection.cursor()

print("Workforce Analytics Database Connected Successfully!")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    Employee_ID INTEGER PRIMARY KEY,
    Employee_Name TEXT NOT NULL,
    Department TEXT NOT NULL,
    Age INTEGER,
    Years_at_Company INTEGER,
    Monthly_Income REAL,
    Job_Satisfaction INTEGER,
    Performance_Rating INTEGER,
    Training_Hours INTEGER,
    Attrition_Status TEXT
)
""")

connection.commit()

print("Employee Table Created Successfully!")
```

```python
query = """
SELECT *
FROM employees;
"""

employee_data = pd.read_sql_query(query, connection)

print("Employee Workforce Data:")
display(employee_data)
```

```python
query = """
SELECT COUNT(*) AS Total_Employees
FROM employees;
"""

total_employees = pd.read_sql_query(query, connection)

print("Total Number of Employees:")
display(total_employees)
```

```python
query = """
SELECT
    Department,
    COUNT(*) AS Employee_Count
FROM employees
GROUP BY Department
ORDER BY Employee_Count DESC;
"""

department_count = pd.read_sql_query(query, connection)

print("Department-wise Employee Count:")
display(department_count)
```

```python
query = """
SELECT
    Department,
    ROUND(AVG(Monthly_Income), 2) AS Average_Monthly_Income
FROM employees
GROUP BY Department
ORDER BY Average_Monthly_Income DESC;
"""

average_income = pd.read_sql_query(query, connection)

print("Average Monthly Income by Department:")
display(average_income)
```

```python
query = """
SELECT
    Department,
    COUNT(*) AS Total_Employees,
    SUM(
        CASE
            WHEN Attrition_Status = 'Yes' THEN 1
            ELSE 0
        END
    ) AS Attrition_Count,
    ROUND(
        100.0 * SUM(
            CASE
                WHEN Attrition_Status = 'Yes' THEN 1
                ELSE 0
            END
        ) / COUNT(*),
        2
    ) AS Attrition_Rate_Percentage
FROM employees
GROUP BY Department
ORDER BY Attrition_Rate_Percentage DESC;
"""

department_attrition = pd.read_sql_query(query, connection)

print("Department-wise Attrition Analysis:")
display(department_attrition)
```

```python
query = """
SELECT
    Employee_ID,
    Employee_Name,
    Department,
    Monthly_Income
FROM employees
ORDER BY Monthly_Income DESC
LIMIT 5;
"""

highest_paid = pd.read_sql_query(query, connection)

print("Top 5 Highest-Paid Employees:")
display(highest_paid)
```

```python
query = """
SELECT
    Department,
    ROUND(AVG(Job_Satisfaction), 2) AS Average_Job_Satisfaction
FROM employees
GROUP BY Department
ORDER BY Average_Job_Satisfaction DESC;
"""

job_satisfaction = pd.read_sql_query(query, connection)

print("Average Job Satisfaction by Department:")
display(job_satisfaction)
```

```python
query = """
SELECT
    Employee_ID,
    Employee_Name,
    Department,
    Years_at_Company,
    Job_Satisfaction,
    Training_Hours,
    Attrition_Status,
    CASE
        WHEN Job_Satisfaction <= 2
             AND Training_Hours < 25 THEN 'High Risk'
        WHEN Job_Satisfaction <= 2
             OR Years_at_Company <= 2 THEN 'Medium Risk'
        ELSE 'Low Risk'
    END AS Risk_Level
FROM employees
WHERE
    Job_Satisfaction <= 2
    OR Training_Hours < 25
    OR Years_at_Company <= 2
ORDER BY
    Job_Satisfaction ASC,
    Training_Hours ASC;
"""

high_risk_employees = pd.read_sql_query(query, connection)

print("Employees with High Attrition Risk Indicators:")
display(high_risk_employees)
```
